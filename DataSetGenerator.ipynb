```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Set random seed for reproducibility
np.random.seed(42)

# Generate integer x values from 1 to 10
x = np.arange(1, 11)

# Define the linear relationship with some noise
# Start with a linear equation: y = mx + c
m = 8  # Slope
c = 30  # Intercept

# Calculate initial y values based on the linear equation
y = m * x + c

# Add some noise to y values to adjust correlation
noise = np.random.normal(0, 10, x.shape)  # Adjust noise level for desired correlation
y = y + noise

# Round y to the nearest integer and ensure they fall within the desired range
y = np.round(y).astype(int)
y = np.clip(y, 20, 100)

# Calculate the correlation coefficient
r, _ = pearsonr(x, y)
print(f"Correlation coefficient (r): {r:.2f}")

# Print the dataset
print("\nGenerated Dataset:")
print("x, y")
for i in range(len(x)):
    print(f"{x[i]}, {y[i]}")

# Plot the data
plt.scatter(x, y)
plt.title('Synthetic Data for Linear Regression')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
