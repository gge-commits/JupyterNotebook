```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Set random seed for reproducibility
np.random.seed(42)

# Generate integer x values from 1 to 10
x = np.arange(1, 11)

# Define the linear relationship with some noise
# Start with a linear equation: y = mx + c
m = 8  # Slope
c = 30  # Intercept

# Calculate initial y values based on the linear equation
y = m * x + c

# Add some noise to y values to adjust correlation
noise = np.random.normal(0, 10, x.shape)  # Adjust noise level for desired correlation
y = y + noise

# Round y to the nearest integer and ensure they fall within the desired range
y = np.round(y).astype(int)
y = np.clip(y, 20, 100)

# Calculate the correlation coefficient
r, _ = pearsonr(x, y)
print(f"Correlation coefficient (r): {r:.2f}")

# Print the dataset
print("\nGenerated Dataset:")
print("x, y")
for i in range(len(x)):
    print(f"{x[i]}, {y[i]}")

# Plot the data
plt.scatter(x, y)
plt.title('Synthetic Data for Linear Regression')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

## House price dataset

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Set random seed for reproducibility
np.random.seed(42)

# Generate family earnings (in thousands)
earnings = np.random.randint(30, 200, 20)  # Earnings between $30k and $200k

# Define a relationship between earnings and house cost
# Assume house cost is approximately 3.5 times the earnings
house_cost = 3.5 * earnings

# Add some variability to house cost to simulate real-world scenarios
noise = np.random.normal(0, 50, earnings.shape)  # Adjust noise level as needed
house_cost = house_cost + noise

# Round house cost to the nearest thousand
house_cost = np.round(house_cost).astype(int)

# Calculate the correlation coefficient
r, _ = pearsonr(earnings, house_cost)
print(f"Correlation coefficient (r): {r:.2f}")

# Print the dataset
print("\nGenerated Dataset:")
print("Earnings (in $1000), House Cost (in $1000)")
for i in range(len(earnings)):
    print(f"{earnings[i]}, {house_cost[i]}")

# Plot the data
plt.scatter(earnings, house_cost)
plt.title('Relationship between Family Earnings and House Cost')
plt.xlabel('Earnings (in $1000)')
plt.ylabel('House Cost (in $1000)')
plt.show()
```
