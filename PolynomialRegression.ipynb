# Polynomial Regression
#### If your data points clearly will not fit a linear regression (a straight line through all data points), it might be ideal for polynomial regression.

#### Polynomial regression, like linear regression, uses the relationship between the variables x and y to find the best way to draw a line through the data points.



#### Polynomial regression is a type of regression analysis where the relationship between the independent variable (x) and the dependent variable (y) is modeled using a polynomial function. Instead of a straight line like in simple linear regression, polynomial regression fits a curve to the data points.

**Key Concepts:**
#### **Polynomial Function:**
A polynomial is an expression with variables and coefficients, where the exponents are non-negative integers (e.g., x, x, x, etc.). 
#### **Nth Degree:**
The "degree" of a polynomial refers to the highest power of the variable in the expression. For example, a quadratic polynomial (degree 2) would have terms like x. 
#### **Non-Linear Relationship:**
Polynomial regression is used when the relationship between the variables is not linear (a straight line) but rather a curve. 
#### **Linear Regression Extension:**
#### Polynomial regression is considered an extension of linear regression because it still fits a model that is linear in the parameters (the coefficients of the polynomial terms), even though the model itself is nonlinear. 
#### **How it works:**
#### **1. Data Input:**
#### You have data points (x, y) where x is the independent variable and y is the dependent variable. 
#### **2. Polynomial Model:**
#### define a polynomial model with a certain degree (e.g., quadratic, cubic). 
#### **3. Fitting the Model:**
#### The model is fitted to the data points by minimizing the sum of squared errors (least squares method). This means finding the polynomial equation that best matches the data. 
#### **4. Prediction:**
#### Once the model is fitted, you can use it to predict the value of y for a given value of x. 
#### Why use polynomial regression? 
#### **Non-linear Relationships:**
#### It's useful when the relationship between the variables cannot be accurately represented by a straight line. 
#### **More Accurate Predictions:**
#### In some cases, polynomial models can provide more accurate predictions than linear models. 
#### **Flexibility:**
#### Polynomial regression allows for a greater degree of flexibility in modeling complex relationships. 
#### **Assumptions of Polynomial Regression:**
#### -  The relationship between the dependent variable and the independent variable is polynomial. 
#### -  The independent variables are independent of each other. 
#### -  There is no multicollinearity (independent variables are not highly correlated). 
#### **Examples:**
#### **Quadratic Regression (Degree 2):** y = b + bx + bx. This model fits a parabola to the data. 
#### **Cubic Regression (Degree 3):** y = b + bx + bx + bx. This model fits a curve with more flexibility.

#### Important Considerations:
#### **Overfitting:**
#### Choosing a polynomial of too high a degree can lead to overfitting, where the model fits the training data very well but performs poorly on new, unseen data.
#### **Model Complexity:**
#### Higher-degree polynomials can be more complex and may be harder to interpret. 
### **Example**
#### Start by drawing a scatter plot

```python
import matplotlib.pyplot as plt

x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]

plt.scatter(x, y)
plt.show()
```

### **Example**
#### Import numpy and matplolib then draw the line of Polynomial Regression


```python
import numpy
import matplotlib.pyplot as plt

x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]

mymodel = numpy.poly1d(numpy.polyfit(x, y, 3)) #NumPy has a method that lets us make a polynomial model:

myline = numpy.linspace(1, 22, 100) #Then specify how the line will display, we start at position 1, and end at position 22:

plt.scatter(x, y) #Draw the scatter plot
plt.plot(myline, mymodel(myline)) #Draw the line using the model we created 
plt.title('Polynomial Regression Example') #Add a title to the plot
plt.xlabel('X-axis') #Label the x-axis
plt.ylabel('Y-axis') #Label the y-axis
plt.show() #Display the plot
```

## R-Squared
#### It is important to know how well the relationship between the values of the x- and y-axis is, if there are no relationship the polynomial regression can not be used to predict anything.

#### The relationship is measured with a value called the r-squared.

#### The r-squared value ranges from 0 to 1, where 0 means no relationship, and 1 means 100% related.

#### Python and the Sklearn module will compute this value for you, all you have to do is feed it with the x and y arrays:

```python
import numpy
from sklearn.metrics import r2_score

x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]

mymodel = numpy.poly1d(numpy.polyfit(x, y, 3))

print(r2_score(y, mymodel(x)))
```

## Predict Future Values

#### To do so, we need the same mymodel array from the example above:

#### Example
#### Predict the speed of a car passing at 17:00:
